```python
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnableSequence, RunnableLambda
```

##Define Base Components

```python
# Prompt template
prompt = ChatPromptTemplate.from_template("Write a short poem about {topic}.")

# Model
model = ChatOpenAI(model="gpt-4o-mini")

# Output parser
parser = StrOutputParser()

# Base sequence
base_seq = RunnableSequence(first=prompt, middle=[model], last=parser)
```

##Add Custom Logic with RunnableLambda

```python
# Preprocess: uppercase the topic
preprocess = RunnableLambda(lambda x: {"topic": x["topic"].upper()})

# Postprocess: add an emoji at the end of the poem
postprocess = RunnableLambda(lambda poem: poem + " 🌟")

# Combine everything in a sequence
custom_chain = preprocess | base_seq | postprocess
```

```python
result = custom_chain.invoke({"topic": "autumn leaves"})
print(result)
```

##Use runnable lambda directly


```python
def poem_func(inputs: dict) -> str:
    topic = inputs["topic"].title()
    prompt_input = {"topic": topic}
    poem = base_seq.invoke(prompt_input)
    return poem + "!"

# Wrap in RunnableLambda
poem_chain = RunnableLambda(poem_func)

result = poem_chain.invoke({"topic": "mountains"})
print(result)
```
