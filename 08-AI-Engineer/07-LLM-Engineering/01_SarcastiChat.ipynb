```python
from google.colab import userdata
import getpass
import os

os.environ["GEMINI_API_KEY"] = userdata.get("GOOGLE_API_KEY")

ngrok_token = getpass.getpass("Enter your ngrok auth token: ")

os.environ["NGROK_AUTH_TOKEN"] = ngrok_token

os.environ["CHAT_DB_PATH"] = "chat_memory.db"
```

```python
from pyngrok import ngrok
import subprocess

# Authenticate ngrok
ngrok.set_auth_token(os.environ["NGROK_AUTH_TOKEN"])

# Start Streamlit in background
port = 8501
streamlit_proc = subprocess.Popen(["streamlit", "run", "app.py", "--server.port", str(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

# Create ngrok tunnel
public_url = ngrok.connect(port).public_url
print(f" Your SarcastiChat is live at: {public_url}")

print(" To stop the app: Runtime > Interrupt execution (or press stop icon)")
```
