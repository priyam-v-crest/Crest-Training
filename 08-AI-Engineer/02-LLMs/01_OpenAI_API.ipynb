## Setup 2

```python
"""
2025-Ready LangChain + Gemini Example
-------------------------------------
Structured, reusable, and safe script to interact with Gemini models
via LangChain. Includes error handling, configurable prompts, and logging.
"""

import os
import logging
from getpass import getpass
from typing import Optional

# LangChain Gemini connector
from langchain_google_genai import ChatGoogleGenerativeAI

# ----------- CONFIGURATION -----------

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s"
)

# Function to securely get API key
def get_api_key() -> str:
    """Prompt user to enter Gemini API key securely."""
    api_key = getpass(" Enter your Gemini API key: ")
    if not api_key:
        logging.error("API key cannot be empty. Exiting.")
        raise ValueError("Missing Gemini API key")
    os.environ["GOOGLE_API_KEY"] = api_key
    return api_key

# ----------- GEMINI INITIALIZATION -----------

def init_gemini_client(api_key: str, model_name: str = "gemini-2.5-flash") -> ChatGoogleGenerativeAI:
    """Initialize the Gemini LLM client via LangChain."""
    try:
        client = ChatGoogleGenerativeAI(model=model_name, google_api_key=api_key)
        logging.info(f"Initialized Gemini client with model: {model_name}")
        return client
    except Exception as e:
        logging.error(f"Failed to initialize Gemini client: {e}")
        raise

# ----------- PROMPT HANDLING -----------

def generate_response(llm_client: ChatGoogleGenerativeAI, prompt: str) -> Optional[str]:
    """Invoke Gemini LLM and return the response content."""
    try:
        response = llm_client.invoke(prompt)
        if hasattr(response, "content"):
            return response.content
        logging.warning("Response object has no 'content' attribute")
        return None
    except Exception as e:
        logging.error(f"Error generating response: {e}")
        return None

# ----------- MAIN FUNCTION -----------

def main():
    # Step 1: Get API key securely
    api_key = get_api_key()

    # Step 2: Initialize LLM client
    llm = init_gemini_client(api_key=api_key)

    # Step 3: Define a creative prompt
    prompt = (
        "You are an expert AI assistant. "
        "Write a futuristic poem imagining LangChain in 2030, "
        "with a mix of humor and technical flair."
    )

    # Step 4: Generate response
    output = generate_response(llm, prompt)

    # Step 5: Print and log the result
    if output:
        print("\n Gemini Response:\n")
        print(output)
    else:
        logging.error("No response generated.")

# Entry point
if __name__ == "__main__":
    main()
```
